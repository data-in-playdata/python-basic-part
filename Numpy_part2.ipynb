````python
import numpy as np

class Calculator:
    val1 = 0
    def __init__(self):  #    val1 default   .
        pass
    def add(self, *val2):
#         if type(val2) is not str and type(val2) is not list and type(val2) is not tuple and type(val2) is not set and type(val2) is not dict and type(val2) is not int and type(val2) is None:     
#             return print(" .")
        if len(val2) == 0:
            return print(" .")
        elif len(val2[0]) > 1:
            for i in val2[0]:
                self.val1 += i
                continue
        else:
            val2 = np.array([val2])
            self.val1 += val2
    def minus(self, *val2):
#         if type(val2) is not str and type(val2) is not list and type(val2) is not tuple and type(val2) is not set and type(val2) is not dict and type(val2) is not int and type(val2) is None:
#             return print(" .")
        if len(val2) == 0:
            return print(" .")
        elif len(val2[0]) > 1:
            for i in val2[0]:
                self.val1 -= i
                continue
        else:
            val2 = np.array([val2])
            self.val1 -= val2
    def show_value(self):
        
        self.show_value = print(f"  {self.val1}.")
````
