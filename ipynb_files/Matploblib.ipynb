# Matplotlib

- 스캐터 플롯(scatter plot)
- 컨투어 플롯(contour plot)
- 서피스 플롯(surface plot)
- 바 차트(bar chart)
- 히스토그램(histogram)
- 박스 플롯(box plot)

### 틱 설정하기
플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고 이 틱에 써진 숫자 혹은 
글자를 틱 라벨(tick label)이라고 합니다. 틱의 위치나 틱 라벨은 맷플롯리브가 자동으로 
정해주지만 만약 수동으로 설정하고 싶다면 xticks()이나 yticks()을 사용합니다.  

````python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)  # -np.pi와 np.pi 사이에 256개의 구간을 균일하게 만든다.
C = np.cos(X) # x값에 맞는 cos값을 만든다.
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
````

틱 라벨 문자열에는 ```$$``` 사이에 수학 문자식(Mathematical expressions)을 넣을 수도 있습니다. 다양한 mathematical expressions은 아래 링크를 통해 확인할 수 있습니다.

````python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)  # -np.pi와 np.pi 사이에 256개의 구간을 균일하게 만든다.
C = np.cos(X) # x값에 맞는 cos값을 만든다.
plt.title("TeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.show()
````

### 여러 개의 선 그리기

스타일 문자열을 반복하여 인수로 넘깁니다. 이 경우에는 하나의 선을 그릴 때 처럼 x 
데이터나 스타일 문자열을 생략할 수 없습니다.  

````python
t = np.arange(0., 5., 0.2)
plt.title('라인 플롯에서 여러 개의 선 그리기')
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
````

하나의 plot()이 아니라 복수의 plot()을 하나의 그림에 겹쳐서 그릴 수도 있다.  

````python
plt.title("복수의 plot 명령을 한 그림에서 표현")
style1 ={'c':'b', 'lw':5, 'ls':'--', 'marker':'o', 'ms':15, 'mec':'g', 'mew':5, 'mfc':'r'}
style2 ={'c':'k', 'lw':3, 'ls':':', 'marker':'s', 'ms':15, 'mec':'m', 'mew':5, 'mfc':'c'}

plt.plot([1,4, 9, 16], **style1)
plt.plot([9, 16, 4, 1], **style2)

plt.show()
````

## Figure
Figure는 전체 그림(그래프)을 의미하는 객체이다. Figure 객체는 Axes, 특수 Artist 그룹(title, figure legends, colorbar, etc), 그리고 심지어 중첩된 subfigures 까지 포함할 수 있다.  
가장 쉬운 방법으로 Figure 객체를 생성하는 방법들을 소개해본다.
1. Axes 없이 Figure만 얻으려면 figure() 메서드를 사용
2. 1개의 Axes 객체를 갖는 Figure를 만들기 위해서 subplots() 메서드를 인수 없이 사용하면 된다.
3. grid 형태의 shape 를 갖는 Axes 객체를 여러 개 만들 땐 subplots() 메서드에 첫 인수로 row 를 두 번째 인수로 column 을 입력한다.

## Axes
Figure 객체 안에 여러 개의 플롯을 grid(마치 표같은) 형태로 보여야 하는 경우가 있다. Figure 안에 있는 각각의 플롯은 Axes 라고 불리는 객체에 속한다.  
Figure 안에 Axes 를 생성하려면 원래 subplot()을 사용해서 명시적으로 Axes 객체를 얻어야 한다. 그러나 plot()을 바로 사용해도 자동으로 Figure, Axes 객체를 생성해준다.  
subplot() 은 그리드(grid) 형태의 Axes 객체들을 생성한다. Figure가 행렬(matrix)이고 Axes가 행렬의 원소라고 생각하면 된다.
위/아래 두 개의 플롯을 하나의 Figure 안에 그리려면 아래처럼 명령을 실행해야 한다. 여기에서 숫자 인덱싱은 파이썬이 Matlab 관행을 따르기 때문에 첫번째 플롯을 가리키는 숫자가 0이 아닌 1인 것에 주의해야 한다.  
tight_layout() 명령을 실행하면 플롯 간의 간격을 자동으로 맞춰준다.  
![image](https://user-images.githubusercontent.com/105477856/216593611-57f2d1f4-7bda-4be6-ad77-35d31e19ce26.png)  
파란색 선으로 감싸둔 곳이 자동으로 맞춰진 간격이다.

````python
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
````

만약 2x2 형태인 네 개의 플롯을 그리려면 Axes 를 4개 형성하면 됩니다. subplot 의 인수는 2,2,1 을 줄여서 221 로 기입해도 괜찮습니다. Axes 의 순서는 콘솔 출력 순서와 동일합니다.(좌에서 우로, 위에서 아래로)  

````python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
````

subplots() 에 인수를 두 개 전달하면 Axes 객체를 동시에 여러 개 생성할 수도 있습니다. 이 때는 2차원 ndarray 형태로 Axes 객체가 반환됩니다.

````python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.show()
````

Axes는 데이터 플로팅을 위한 영역으로 Figure에 포함되는 Artist입니다. 일반적으로 2차원의 경우 Axis 객체를 두 개 포함(3차원의 경우 Axis 객체를 세 개 포함)합니다. Axis는 ticks와 tick labels를 제공하며 이는 Axes의 데이터 범위에 대한 기준이 됩니다.  
- Axes와 Axis는 다른 객체임을, 헷갈리지 않도록 합시다.  
Axes 클래스와 그 메서드는 주요한 Entry point이며, 대부분의 플로팅 메서드가 정의되어 있습니다.

### 컨투어 플롯
입력 변수가 x, y 로 두 개 이고 출력 변수가 z 하나인 경우에 3차원 자료가 됩니다.  
이런 데이터를 시각화하는 방법으로 명암이 아닌 등고선(contour)을 사용할 수 있습니다.    
그러기 위해 등고선만을 표시하는 contour와 색까지 표시하는 contourf 명령을 사용할 수 있습니다.  
입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 합니다.  

````python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
````

### 3D 서피스 플롯
3차원 플롯은 등고선 플롯과 달리 Axes3D 라는 3차원 전용 Axes를 생성해서 입체적으로 표현합니다.  
plot_wireframe(), plot_surface() 명령을 사용합니다.

````python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
````
